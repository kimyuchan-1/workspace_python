```python
"""곱셈 표 출력"""

upper = int(input('1부터 몇까지 곱할까요：'))

def multiplication_table(n: int) -> bool:
    """1 ~ n까지의 곱셈 표를 출력"""
    if    1 <= n <=  3: w = 2
    elif  4 <= n <=  9: w = 3
    elif 10 <= n <= 31: w = 4
    else              : return False

    f = '{{:{}d}}'.format(w) # 출력 format 설정: 입력받은 숫자에 따라 출력 간격을 설정 ex) :3d
    print('-' * n * w)
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            print(f.format(i * j), end='')
        print()
    print('-' * n * w)
    return True

multiplication_table(upper)
```

```python
"""곰셈 표 출력(내부 함수)"""

upper = int(input('1부터 몇까지 곱할까요：'))

def multiplication_table(n: int) -> bool:
    """1 ~ n까지의 곱셈 표를 출력"""

    def put_bar(n: int) -> None:
        """n개의 '-'을 연속해서 출력하고 개행"""
        print('-' * n)
        n=0
        print(id(n))

    if    1 <= n <=  3: w = 2
    elif  4 <= n <=  9: w = 3
    elif 10 <= n <= 31: w = 4
    else              : return False

    f = '{{:{}d}}'.format(w)
    put_bar(n * w)
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            print(f.format(i * j), end='')
        print()
    put_bar(n * w)
    print(id(n))
    return True

multiplication_table(upper)
```

```python
"""곱셈 표 출력(내부 함수 : 인수 삭제)"""

upper = int(input('1부터 몇까지 곱할까요：'))

def multiplication_table(n: int) -> bool:

    """1 ~ n까지의 곱셈 표를 출력"""
    if    1 <= n <=  3: w = 2
    elif  4 <= n <=  9: w = 3
    elif 10 <= n <= 31: w = 4
    else              : return False

    def put_bar() -> None:
        """n * w개의 '-'을 연속해서 출력하고 개행"""
        print('-' * n * w)
        print(id(n))

    f = '{{:{}d}}'.format(w)
    put_bar()
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            print(f.format(i * j), end='')
        print()
    put_bar()
    print(id(n))
    return True

multiplication_table(upper)
```

```python
"""변수 내외의 변수 값을 확인"""

g1 = 1
g2 = 2

def outer():
    def inner():
        f2 = 6
        print('[2] inner() g1, g2 = ', g1, g2)
        print('    inner() f1, f2 = ', f1, f2)

    f1 = 3
    f2 = 4
    g2 = 5
    inner()
    print('[3] outer() g1, g2 = ', g1, g2)
    print('    outer() f1, f2 = ', f1, f2)

print('[1] global  g1, g2 = ', g1, g2)
outer()
print('[4] global  g1, g2 = ', g1, g2)
```

```python
# global문 있음

n = 1

def func():
    global n
    n = 2    # n은 글로벌 변수
    print('n =', n)

print('n =', n)
func()
print('n =', n)
```

```python
# nonlocal문 있음

def outer():
    n = 1
    def inner():
        nonlocal n
        n = 2
        print('n =', n)

    print('n =', n)
    inner()
    print('n =', n)

outer()
```
