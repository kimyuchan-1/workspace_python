```python
# 어노테이션이 있는 함수 puts

def puts(n: int, s: str) -> None:
    """n개의 s를 연속해서 출력"""
    for _ in range(n):
        print(s, end='')

puts(5, '*')
print()
print(puts.__annotations__)
print()
puts('*', 5)
```

```python
"""어노테이션과 문서화 문자열이 있는 함수 puts"""


def puts(n: int, s: str) -> None:
    """n개의 s를 연속해서 출력

    매개 변수:
        n -- 출력하는 문자열의 개수
        s -- 출력하는 문자열
    리턴값:
        없음

    """
    for _ in range(n):
        print(s, end='')


print(puts.__doc__)    # 문서화 문자열-doc string을 출력
```

```python
"""특정 해의 일수 구하기"""

def is_leapyear(year: int) -> bool:
    """서기 year년은 윤년인지 알아보기"""
    return year % 4 == 0 and year % 100 != 0 or year % 400 == 0

print('특정 해의 일수를 구합니다.')
y = int(input('구할 년도：'))
print('그 해는 {}일입니다.'.format(365 + is_leapyear(y)))
```

```python
"""딕셔너리에서 특정한 값을 가지는 키의 리스트 생성"""

def keys_of(dic: dict, val: 'value') -> list:
    """딕셔너리 dic에 있는 값이 val인 요소의 키를 리스트로 리턴"""
    return [k for k, v in dic.items() if v == val]

txt = input('문자열 : ')
count = {ch: txt.count(ch) for ch in txt}
print('분포=', count)

num = int(input('몇 개짜리 문자：'))
print('{}개의 문자={}'.format(num, keys_of(count, num)))
```
