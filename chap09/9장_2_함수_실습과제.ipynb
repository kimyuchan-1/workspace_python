Ï†êÏàò Î¶¨Ïä§Ìä∏Î•º Î∞õÏïÑ 70Ï†ê Ïù¥ÏÉÅÏù¥Î©¥ 'Ìï©Í≤©', Í∑∏Î†áÏßÄ ÏïäÏúºÎ©¥ 'Î∂àÌï©Í≤©'ÏúºÎ°ú Î∂ÑÎ•òÎêú Î¨∏ÏûêÏó¥ Î¶¨Ïä§Ìä∏Î•º Î∞òÌôòÌïòÏÑ∏Ïöî.

Ï∂úÎ†•: ['Ìï©Í≤©', 'Î∂àÌï©Í≤©', 'Ìï©Í≤©', 'Î∂àÌï©Í≤©']


```python
def classify_scores(scores):
    return ['Ìï©Í≤©' if s >= 70 else 'Î∂àÌï©Í≤©' for s in scores]

scores = [85, 67, 90, 45]
classify_scores(scores)
```

üìå Í≥ºÏ†ú 2

Ïù¥Î¶Ñ Î¶¨Ïä§Ìä∏Î•º Î∞õÏïÑ Í∞Å Ïù¥Î¶ÑÏóê ÏàúÎ≤àÏùÑ Î∂ôÏù∏ Î¶¨Ïä§Ìä∏Î•º Î∞òÌôòÌïòÏÑ∏Ïöî (enumerate() ÏÇ¨Ïö©).

Ï∂úÎ†• ÏòàÏãú: ['1. ÌôçÍ∏∏Îèô', '2. ÍπÄÏú†Ïã†', ...]

```python
def numbered_names(names):
    return [f"{i+1}. {name}" for i, name in enumerate(names)]

names = ["ÌôçÍ∏∏Îèô", "ÍπÄÏú†Ïã†", "Í∞ïÍ∞êÏ∞¨"]
numbered_names(names)
```

```python
def combine_name_score(names, scores):
    return [f"{name}: {score}Ï†ê" for name, score in zip(names, scores)]

combine_name_score(names, scores)
```

Îã§ÏàòÏùò (Ïù¥Î¶Ñ, Ï†êÏàò) ÌäúÌîå Î¶¨Ïä§Ìä∏Î•º Î∞õÏïÑ, ÌèâÍ∑† Ï†êÏàòÏôÄ ÏµúÍ≥† Ï†êÏàò ÌïôÏÉùÏùÑ Ï∂úÎ†•ÌïòÏÑ∏Ïöî.

üíé  STEP 6: ÌôïÏû• Í≥ºÏ†ú (Ïã§Ï†ÑÌòï)

üìå Í≥ºÏ†ú 10: Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨ ‚Üí Î∂ÑÏÑùÍπåÏßÄ Ï†Ñ Í≥ºÏ†ï Ìï®Ïàò Î∂ÑÎ¶¨ Ïã§Ïäµ

ÏïÑÎûòÏùò Î¶¨Ïä§Ìä∏Î•º Í∏∞Î∞òÏúºÎ°ú Îã§Ïùå ÏûëÏóÖÏùÑ Í∞ÅÍ∞ÅÏùò Ìï®ÏàòÎ°ú Î∂ÑÎ¶¨ÌïòÏó¨ Íµ¨ÌòÑÌï¥Î≥¥ÏÑ∏Ïöî.

names = ["ÌôçÍ∏∏Îèô", " ÍπÄÏú†Ïã† ", "Ïù¥ÏàúÏã†", "Í∞ï Í∞ê Ï∞¨"]

scores = [85, 70, 60, 90]

subjects = ["ÏàòÌïô", "ÏòÅÏñ¥", "Í≥ºÌïô", "Íµ≠Ïñ¥"]

Ìï®Ïàò Î∂ÑÎ¶¨ Î™©Ìëú

clean_names() ‚Üí Ïù¥Î¶ÑÏóêÏÑú Í≥µÎ∞± Ï†úÍ±∞ Î∞è Ï†ïÎ¶¨

to_dict() ‚Üí Ïù¥Î¶ÑÍ≥º Ï†êÏàò ÎîïÏÖîÎÑàÎ¶¨Î°ú Î≥ÄÌôò

filter_pass() ‚Üí 70Ï†ê Ïù¥ÏÉÅÏù∏ ÌïôÏÉùÎßå ÌïÑÌÑ∞ÎßÅ

report() ‚Üí "1. ÌôçÍ∏∏Îèô - ÏàòÌïô: 85Ï†ê" ÌòïÌÉúÎ°ú Ï∂úÎ†•

```python
def clean_names(names: list) -> list:
    """Ïù¥Î¶ÑÏóêÏÑú Í≥µÎ∞± Ï†úÍ±∞ Î∞è Ï†ïÎ¶¨"""
    return [name.strip().replace(" ","") for name in names]

def to_dict(names: list, scores: list) -> dict:
    """Ïù¥Î¶ÑÍ≥º Ï†êÏàò ÎîïÏÖîÎÑàÎ¶¨Î°ú Î≥ÄÌôò"""
    return {name:score for name, score in zip(names, scores)}

def filter_pass(data: dict) -> list:
    """70Ï†ê Ïù¥ÏÉÅÏù∏ ÌïôÏÉùÎßå ÌïÑÌÑ∞ÎßÅ"""
    return [(key, value) for key, value in data.items() if value >= 70]

def report(filteredData: list, subjects: list) -> None:
    """1. ÌôçÍ∏∏Îèô - ÏàòÌïô: 85Ï†ê" ÌòïÌÉúÎ°ú Ï∂úÎ†•"""
    for i, ((name, score), subject) in enumerate((zip(filteredData, subjects)), start=1):
        print(f'{i}. {name} - {subject}: {score}Ï†ê' )

names=["ÌôçÍ∏∏Îèô", " ÍπÄÏú†Ïã† ", "Ïù¥ÏàúÏã†", "Í∞ï Í∞ê Ï∞¨"]
scores = [85, 70, 60, 90]
subjects = ["ÏàòÌïô", "ÏòÅÏñ¥", "Í≥ºÌïô", "Íµ≠Ïñ¥"]

cleanNames = clean_names(names)
print(cleanNames)
nsDictionary = to_dict(cleanNames, scores)
print(nsDictionary)
filteredData = filter_pass(nsDictionary)
print(filteredData)
report(filteredData, subjects)

```

Î¶¨Ïä§Ìä∏ÏóêÏÑú Í∞ÄÏû• ÎßéÏù¥ Îì±Ïû•ÌïòÎäî ÏöîÏÜåÎ•º Î∞òÌôòÌïòÎäî Ìï®Ïàò (ÎπàÎèÑÏàò Î∂ÑÏÑù)

ÌïôÏÉù Ï†êÏàò ÎîïÏÖîÎÑàÎ¶¨Î•º Î∞õÏïÑ 'Ìï©Í≤©/Î∂àÌï©Í≤©' ÎîïÏÖîÎÑàÎ¶¨ Î∞òÌôò (Í∏∞Ï§Ä 70Ï†ê)

```python
def rangeList(scores: list) -> list: 
    """Ïà´Ïûê Î¶¨Ïä§Ìä∏ÏóêÏÑú ÌäπÏ†ï Î≤îÏúÑ(Ïòà: 50~80)Ïóê Ìï¥ÎãπÌïòÎäî Í∞íÎßå Î∞òÌôò"""
    return [score for score in scores if 50<=score<=80]
```

```python
import random

def theMostElementinList(nList: list) -> list:
    """Î¶¨Ïä§Ìä∏ÏóêÏÑú Í∞ÄÏû• ÎßéÏù¥ Îì±Ïû•ÌïòÎäî ÏöîÏÜåÎ•º Î∞òÌôòÌïòÎäî Ìï®Ïàò (ÎπàÎèÑÏàò Î∂ÑÏÑù)"""
    freq = {}
    for i in nList:
        freq[i] = freq.get(i,0) + 1
    
    maxCount = max(freq.values())
    return [key for key, value in freq.items() if value == maxCount ]

nList = [random.randint(1,100) for _ in range(10)]
modes = theMostElementinList(nList)
print(modes)
```

‚úÖ Î≥¥Ï∂© STEP 2: Ïã§Ïö© Îç∞Ïù¥ÌÑ∞ Ï≤òÎ¶¨ Ìï®Ïàò (ML ÏπúÌôîÏ†Å)

üéØ Î™©Ìëú

Î®∏Ïã†Îü¨ÎãùÏóêÏÑú ÏûêÏ£º Ïì∞Îäî Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨Ïö© Ìï®Ïàò Íµ¨Ï°∞ ÏùµÌûàÍ∏∞

üìù Ïã§Ïäµ Í≥ºÏ†ú

Í≤∞Ï∏°Í∞í(NaN ÎòêÎäî None)Ïù¥ ÏûàÎäî Îç∞Ïù¥ÌÑ∞Î•º ÌïÑÌÑ∞ÎßÅÌïòÎäî Ìï®Ïàò

Î¨∏ÏûêÏó¥ Î¶¨Ïä§Ìä∏Î•º Ïà´ÏûêÎ°ú Î∞îÍæ∏Îäî Îß§Ìïë Ìï®Ïàò (['ÎÇ®', 'Ïó¨'] ‚Üí [0,1])

Ïó¨Îü¨ Í∞úÏùò Ïà´Ïûê Î¶¨Ïä§Ìä∏Î•º Î∞õÏïÑ Ï†ïÍ∑úÌôîÎêú Î¶¨Ïä§Ìä∏(0~1)Î°ú Î∞òÌôò

```python
def filter_NaN(data: list) -> list:
    """Í≤∞Ï∏°Í∞í(NaN ÎòêÎäî None)Ïù¥ ÏûàÎäî Îç∞Ïù¥ÌÑ∞Î•º ÌïÑÌÑ∞ÎßÅÌïòÎäî Ìï®Ïàò"""
    return [i for i in data if i != None]
```

```python
def mapping_gender(data: list) -> list:
    """Î¨∏ÏûêÏó¥ Î¶¨Ïä§Ìä∏Î•º Ïà´ÏûêÎ°ú Î∞îÍæ∏Îäî Îß§Ìïë Ìï®Ïàò (['ÎÇ®', 'Ïó¨'] ‚Üí [0,1])"""
    return [0 if s == "ÎÇ®" else 1 for s in data]
```

```python
def normalize(data: list) -> list:
    """Ïó¨Îü¨ Í∞úÏùò Ïà´Ïûê Î¶¨Ïä§Ìä∏Î•º Î∞õÏïÑ Ï†ïÍ∑úÌôîÎêú Î¶¨Ïä§Ìä∏(0~1)Î°ú Î∞òÌôò"""
    min_val = min(data)
    max_val = max(data)
    return [(x - min_val) / (max_val - min_val) for x in data]
```

‚úÖ Î≥¥Ï∂© STEP 3: Î∂ÑÏÑù Î≥¥Í≥† Ìï®Ïàò

üéØ Î™©Ìëú

Îç∞Ïù¥ÌÑ∞Î•º Î∞õÏïÑ ÏöîÏïΩ Ï†ïÎ≥¥, ÌÜµÍ≥Ñ Í≤∞Í≥º, ÏãúÍ∞ÅÌôî Îì± Î¶¨Ìè¨Ìä∏ÏÑ± Ï∂úÎ†• ÏÉùÏÑ±

dict ÌòïÌÉúÏùò Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Ìï≠Î™©Î≥Ñ ÎπàÎèÑÏàòÎ•º Ï∂úÎ†•ÌïòÎäî Ìï®Ïàò

Ï£ºÏñ¥ÏßÑ Î≤îÏ£ºÌòï Îç∞Ïù¥ÌÑ∞Î•º Î∞õÏïÑ ÎßâÎåÄÍ∑∏ÎûòÌîÑÎ•º Í∑∏Î¶¨Îäî Ìï®Ïàò (matplotlib Ïù¥Ïö©)

```python
import random

def max_min_avg_sd(nList: list) -> tuple:
    """Ïà´Ïûê Î¶¨Ïä§Ìä∏Î•º Î∞õÏïÑ ÏµúÎåÄ, ÏµúÏÜå, ÌèâÍ∑†, Ï§ëÏïôÍ∞í, ÌëúÏ§ÄÌé∏Ï∞®Î•º Î∞òÌôòÌïòÎäî Ìï®Ïàò"""
    maxN = max(nList)
    minN = min(nList)
    avgN = sum(nList) / len(nList)
    sdN = (sum((x - avgN) **2 for x in nList ) / len(nList)-1) ** 0.5
    return maxN, minN, avgN, sdN

nList = random.sample(range(1, 101), 10)
print(nList)
print(max_min_avg_sd(nList))
```

```python
import string
import random

def printFrequency(sDict: dict) -> None:
    """dict ÌòïÌÉúÏùò Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Ìï≠Î™©Î≥Ñ ÎπàÎèÑÏàòÎ•º Ï∂úÎ†•ÌïòÎäî Ìï®Ïàò"""
    freq = {}

    for i in sDict.values():
        freq[i] = freq.get(i, 0) + 1

    for value, cnt in sorted(freq.items()):
        print(f'{value}:{cnt}')

keys = random.sample(string.ascii_lowercase, 10)
sDict = {random.randint(1,10) : k for k in keys}
print(sDict)

print(printFrequency(sDict))
```

üíé Î≥¥Ï∂© STEP 4: Ïã§Ï†ÑÌòï Ìï®Ïàò ÌõàÎ†®

üéØ Î¨∏Ï†ú:

CSVÏóêÏÑú Îç∞Ïù¥ÌÑ∞Î•º ÏùΩÍ≥†, Í≤∞Ï∏°ÏπòÎ•º Ï†úÍ±∞ÌïòÍ≥†, Ï†ïÍ∑úÌôîÎ•º Ï†ÅÏö©Ìïú ÌõÑ, ÌèâÍ∑† Ïù¥ÏÉÅÏù∏ Îç∞Ïù¥ÌÑ∞Îßå ÌïÑÌÑ∞ÎßÅÌïòÎäî ÏùºÎ†®Ïùò Í≥ºÏ†ïÏùÑ Ìï®Ïàò Îã®ÏúÑÎ°ú ÎÇòÎàÑÏñ¥ Íµ¨ÌòÑÌïòÏÑ∏Ïöî.

```python
import csv

def read_csv_as_dicts(filename):

    rows = []

    with open(filename, "r", encoding="utf-8-sig") as f:
        reader = csv.DictReader(f)

        for row in reader:
            rows.append(row)

    return rows

def delNaN(data):
    return [ item for item in data if "" not in item.values()]

def normalization(data):
    aList = [int(item["age"]) for item in data]
    hList = [int(item["height"]) for item in data]
    wList = [int(item["weight"]) for item in data]
    min_a, max_a = min(aList), max(aList)
    min_h, max_h = min(hList), max(hList) 
    min_w, max_w = min(wList), max(wList)

    for item in data:
        item["age"] = round((int(item["age"]) - min_a) / (max_a - min_a), 2)
        item["height"] = round((int(item["height"]) - min_h) / (max_h - min_h),2)
        item["weight"] = round((int(item["weight"]) - min_w) / (max_w - min_w),2)

    return data

def filtering_data(data):
    return [ item for item in data if item["age"] >= 0.5 and item["height"] >= 0.5 and item["weight"] >= 0.5]

# ÏÇ¨Ïö© ÏòàÏãú
data = read_csv_as_dicts("sample_data.csv")

print("Raw data")
for item in data:
    print(item)
print()

delete_NaN_data = delNaN(data)

print("Delete NaN in data")
for item in delete_NaN_data:
    print(item)
print()

normalized_data = normalization(delete_NaN_data)

print("Normalized data")
for item in normalized_data:
    print(item)
print()

filtered_data = filtering_data(normalized_data)

print("Filtered data")
for item in filtered_data:
    print(item)
print()
```
