```python
def clean_names(names: list) -> list:
    """ì´ë¦„ì—ì„œ ê³µë°± ì œê±° ë° ì •ë¦¬"""
    return [name.strip().replace(" ","") for name in names]

def to_dict(names: list, scores: list) -> dict:
    """ì´ë¦„ê³¼ ì ìˆ˜ ë”•ì…”ë„ˆë¦¬ë¡œ ë³€í™˜"""
    return {name:score for name, score in zip(names, scores)}

def filter_pass(data: dict) -> list:
    """70ì  ì´ìƒì¸ í•™ìƒë§Œ í•„í„°ë§"""
    return [(key, value) for key, value in data.items() if value >= 70]

def report(filteredData: list, subjects: list) -> None:
    """1. í™ê¸¸ë™ - ìˆ˜í•™: 85ì " í˜•íƒœë¡œ ì¶œë ¥"""
    for i, ((name, score), subject) in enumerate((zip(filteredData, subjects)), start=1):
        print(f'{i}. {name} - {subject}: {score}ì ' )

names=["í™ê¸¸ë™", " ê¹€ìœ ì‹  ", "ì´ìˆœì‹ ", "ê°• ê° ì°¬"]
scores = [85, 70, 60, 90]
subjects = ["ìˆ˜í•™", "ì˜ì–´", "ê³¼í•™", "êµ­ì–´"]

cleanNames = clean_names(names)
print(cleanNames)
nsDictionary = to_dict(cleanNames, scores)
print(nsDictionary)
filteredData = filter_pass(nsDictionary)
print(filteredData)
report(filteredData, subjects)

```

```python
import random

def theMostElementinList(nList: list) -> list:
    """ë¦¬ìŠ¤íŠ¸ì—ì„œ ê°€ì¥ ë§ì´ ë“±ì¥í•˜ëŠ” ìš”ì†Œë¥¼ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜ (ë¹ˆë„ìˆ˜ ë¶„ì„)"""
    freq = {}
    for i in nList:
        freq[i] = freq.get(i,0) + 1
    
    maxCount = max(freq.values())
    return [key for key, value in freq.items() if value == maxCount ]

nList = [random.randint(1,100) for _ in range(10)]
modes = theMostElementinList(nList)
print(modes)
```
