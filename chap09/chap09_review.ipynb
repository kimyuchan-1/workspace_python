```python
def p(n):
    if n == 0: return 1
    return p(n-1) * n

p(5)
```

```python
def r(s):
    if len(s) == 1: return s[0]
    return s[-1] + r(s[1:])

r("hello")
```

```python
def is_leap(year: int) -> bool:
    return "윤년" if year % 4 == 0 and year % 100 != 0 or year % 400 == 0 else "윤년 아님"

is_leap(2000)
```
