```python
class arange:
    def __init__(self, n):
        self.n = n
        self.curr = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.curr >= self.n:
            raise StopIteration()
        curr = self.curr
        self.curr += 1
        return curr

    def __call__(self):
        return self.__next__()
```

```python
for i in arange(5):
    print(i)
```

```python
nums = list(arange(5))
nums
```
