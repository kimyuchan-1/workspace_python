```python
# 510ÌéòÏù¥ÏßÄ - ÎÑàÎ¨¥ Ïñ¥Î†µÎã§ > ÏÇ¨Ïö©ÌïòÏßÄ ÏïäÏùå
def deco(func):
    def wrapper(*args, **kwargs):
        print(f"[LOG] Ìï®Ïàò {func.__name__} Ìò∏Ï∂ú")
        result = func(*args, **kwargs)
        print(f"[LOG] Ìï®Ïàò {func.__name__} Ï¢ÖÎ£å")
        return result
    return wrapper
### func.__name__ ÏùÄ Ìï®ÏàòÍ∞Ä Ï†ïÏùòÎê† Îïå PythonÏù¥ ÏûêÎèôÏúºÎ°ú ÏÑ§Ï†ïÌïòÎäî Ìï®Ïàò Í∞ùÏ≤¥Ïùò ÏÜçÏÑ±
```

```python
class Member:
    def __init__(self, name, weight):
        self._name = name # Ïñ∏ÎçîÏä§ÏΩîÏñ¥ ÌïòÎÇò(_)Îäî "Ïô∏Î∂ÄÏóêÏÑú ÏßÅÏ†ë Ï†ëÍ∑ºÌïòÏßÄ ÎßêÏïÑ Ï£ºÏÑ∏Ïöî"ÎùºÎäî Í¥ÄÎ°ÄÏ†Å ÏùòÎØ∏
        ## ÌååÏù¥Ïç¨Ïù¥ Ï†ëÍ∑ºÏùÑ ÎßâÏïÑÏ£ºÎäî Í∏∞Îä•ÏùÄ ÏóÜÎã§.
        self._weight = weight

    # -------- getter Ìï®Ïàò --------
    def get_weight(self):
        print("‚öô getter Ìò∏Ï∂ú")
        return self._weight

    # -------- setter Ìï®Ïàò --------
    def set_weight(self, w):
        print("‚öô setter Ìò∏Ï∂ú")
        if w > 0:
            self._weight = w
        else:
            self._weight = 0

    # -------- deleter Ìï®Ïàò --------
    def del_weight(self):
        print("‚öô deleter Ìò∏Ï∂ú")
        del self._weight # _weight Î≥ÄÏàòÎäî ÏùòÎèÑÍ∞Ä private Ïö©ÎèÑÏù¥ÎÇò Ï†úÌïúÏùÑ ÎëêÏßÄ ÏïäÎäîÎã§-Îã®Ïàú Ï∞∏Í≥†Ïö©

    # weight ÌîÑÎ°úÌçºÌã∞ ÏÉùÏÑ± > _weight ÎåÄÏã†Ïóê weightÎ•º ÏÇ¨Ïö©ÌïúÎã§Îäî ÏùòÎØ∏
    weight = property(get_weight, set_weight, del_weight, doc="Ï≤¥Ï§ë ÌîÑÎ°úÌçºÌã∞")
```

```python
class Member:
    def __init__(self, name, weight):
        self._name = name
        self._weight = weight

    @property
    def weight(self):
        print("‚öô getter Ìò∏Ï∂ú")
        return self._weight

    @weight.setter
    def weight(self, w):
        print("‚öô setter Ìò∏Ï∂ú")
        self._weight = w if w > 0 else 0

    @weight.deleter
    def weight(self):
        print("‚öô deleter Ìò∏Ï∂ú")
        del self._weight
```

ÌååÏù¥Ïç¨Ïùò __str__ Í≥º __repr__ Ïùò Ï∞®Ïù¥Î•º Í∞ÄÏû• Î™ÖÌôïÌïòÍ≤å Î≥¥Ïó¨Ï£ºÎäî ÏòàÏ†ú

```python
class Member:
    def __init__(self, no, name, weight):
        self.no = no
        self.name = name
        self.weight = weight

    def __str__(self):
        return f"{self.name} ÌöåÏõê ({self.weight}kg)"   # ÏÇ¨Ïö©Ïûê ÏπúÌôîÏ†Å

    def __repr__(self):
        return f"Member(no={self.no}, name='{self.name}', weight={self.weight})"  # Í∞úÎ∞úÏûê ÏπúÌôîÏ†Å
```

Ïù¥Î•º ‚Äúrepr Ïû¨ÌòÑ Í∞ÄÎä•(reproducible, unambiguous) ÌëúÌòÑ‚Äù Ïù¥ÎùºÍ≥† Î∂ÄÎ•∏Îã§


```python
class Member:
    def __init__(self, no, name, weight):
        self.no = no
        self.name = name
        self.weight = weight

    def __repr__(self):
        # Ïù¥ Î¨∏ÏûêÏó¥Ïù¥ Member Í∞ùÏ≤¥ ÏÉùÏÑ±Ïãù Í∑∏ÎåÄÎ°úÏó¨Ïïº ÌïúÎã§
        return f"Member({self.no!r}, {self.name!r}, {self.weight!r})"
        # !rÏùÄ __repr()Ïä§ÌÉÄÏùº Ï†ÅÏö©ÌïòÎùºÎäî ÏùòÎØ∏
```

```python
kim = Member(15, "ÍπÄÏ§ÄÌò∏", 72.7)

text = repr(kim) #repr()Ïù¥ **‚ÄúÌï¥Îãπ Í∞ùÏ≤¥Î•º Îã§Ïãú ÎßåÎì§ Ïàò ÏûàÎäî ÌòïÌÉúÏùò Î¨∏ÏûêÏó¥‚Äù**ÏùÑ Ï†úÍ≥µ
print("repr =", text)

obj = eval(text) #eval() ÏùÄ Î¨∏ÏûêÏó¥ÏùÑ Python ÌëúÌòÑÏãùÏ≤òÎüº Ïã§ÌñâÌïòÏó¨ Í∑∏ Í∞íÏùÑ Î∞òÌôòÌïòÎäî Ìï®Ïàò
## eval()Ïù¥ Í∑∏ Î¨∏ÏûêÏó¥ÏùÑ Ïã§Ï†ú ÏΩîÎìúÏ≤òÎüº Ïã§ÌñâÌï¥ Í∞ùÏ≤¥Î•º ÏÉùÏÑ± - Í∞ùÏ≤¥ ÏãùÎ≥ÑÏûê + Ï∞∏Ï°∞Î≥ÄÏàòÎ°úÏÑú Í∞ùÏ≤¥Î•º Í∞ÄÎ¶¨ÏºúÏïº ÌïúÎã§
print("Î≥µÏõêÎêú Í∞ùÏ≤¥ =", obj)
print("ÎèôÏùºÌïúÍ∞Ä? ‚Üí", isinstance(obj, Member))
```

```python
class Member:
    def __init__(self, no, name, tags):
        self.no = no
        self.name = name
        self.tags = tags

    def __repr__(self):
        return f"Member({self.no!r}, {self.name!r}, {self.tags!r})"

m1 = Member(10, "Alice", ["Ìó¨Ïä§", "ÏöîÍ∞Ä"])
copy = eval(repr(m1))

print(m1)
print(copy)
print(m1.tags == copy.tags)   # True
print(m1 is copy)             # False Î≥ÑÍ∞ú Í∞ùÏ≤¥
```

```python
class Member:
    total_count = 0   # ÌÅ¥ÎûòÏä§ Î≥ÄÏàò (Î™®Îì† Í∞ùÏ≤¥Í∞Ä Í≥µÏú†)

    def __init__(self, name):
        self.name = name # Í∞ùÏ≤¥ Î≥ÄÏàò(Ïù∏Ïä§ÌÑ¥Ïä§ Î≥ÄÏàò)
        Member.total_count += 1

    # Ïù∏Ïä§ÌÑ¥Ïä§ Î©îÏÜåÎìú ‚Üí selfÎ°ú Í∞úÎ≥Ñ Í∞ùÏ≤¥Ïóê Ï†ëÍ∑º
    def hello(self):
        print(f"ÏïàÎÖïÌïòÏÑ∏Ïöî, Ï†ÄÎäî {self.name}ÏûÖÎãàÎã§.")

    # ÌÅ¥ÎûòÏä§ Î©îÏÜåÎìú ‚Üí clsÎ°ú ÌÅ¥ÎûòÏä§ ÏûêÏ≤¥Ïóê Ï†ëÍ∑º
    @classmethod # 1Î≤à ÏÇ¨Ìï≠(Ïñ¥ÎÖ∏ÌÖåÏù¥ÏÖò)
    def show_total(cls):
        print(f"ÌöåÏõê Ïàò: {cls.total_count}Î™Ö") # 2Î≤à ÏÇ¨Ìï≠(clsÎ°ú Ï†ëÍ∑º)

    # Ï†ïÏ†Å Î©îÏÜåÎìú ‚Üí ÌÅ¥ÎûòÏä§/Í∞ùÏ≤¥ÏôÄ Î¨¥Í¥Ä, Ïú†Ìã∏Î¶¨Ìã∞ Ìï®Ïàò
    @staticmethod
    def is_valid_name(name):
        return len(name) >= 2
```

```python
print(f'{Member.total_count}')
print(f'{Member.show_total()}')
```

```python
kim = Member("ÍπÄÏ§ÄÌò∏")
lee = Member("Ïù¥ÏàòÏó∞")
print(f'{lee.total_count}')
kim.hello()            # Ïù∏Ïä§ÌÑ¥Ïä§ Î©îÏÜåÎìú
Member.show_total()    # ÌÅ¥ÎûòÏä§ Î©îÏÜåÎìú
print(Member.is_valid_name("Ìôç"))   # Ï†ïÏ†Å Î©îÏÜåÎìú
print(Member.is_valid_name("A"))    # Ï†ïÏ†Å Î©îÏÜåÎìú
```

```python
class MyList:
    def __init__(self,data):
        self.data = data

    def __getitem__(self, index):
        return self.data[index]
    # Î¶¨Ïä§Ìä∏Ï≤òÎüº obj[index] ÎòêÎäî obj[start:stop] Í∞ôÏùÄ Ïù∏Îç±Ïã±/Ïä¨ÎùºÏù¥Ïã± Ïó∞ÏÇ∞ÏùÑ
    # ÏÇ¨Ïö©Ìï† Îïå ÏûêÎèôÏúºÎ°ú Ìò∏Ï∂úÎêòÎäî Î©îÏÑúÎìú
mlist = MyList([1,2,3,4])
print(mlist[2:]) # Í∞ùÏ≤¥Ïóê ÎåÄÌïú Ïù∏Îç±Ïã± -> __getitem__() ÏûêÎèô Ìò∏Ï∂ú
```
