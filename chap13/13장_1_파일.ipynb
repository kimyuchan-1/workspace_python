```python
# 파일에 문자열 두 줄 입력하기

f = open('hello.txt', 'w')      # 열기(텍스트 + 쓰기 모드)

f.write('Hello!\n')
f.write('How are you?\n')
f.write("good days\n")

f.close()                       # 닫기
```

```python
# 파일에서부터 두 줄의 문자열을 읽어 들여 출력하기

f = open('hello.txt', encoding= "UTF-8")       # 열기(텍스트 + 쓰기 모드)

line1 = f.readline()        # 한 줄 읽어 들이기
line2 = f.readline()        # 한 줄 읽어 들이기
line3 = f.readline()   
print(line1, end='')
print(line2, end='')
print(line3, end='')
f.close()                   # 닫기
```

```python
# 파일에 두 줄의 문자열을 추가하기

f = open('hello.txt', 'a')  # 열기(텍스트 + 추가 쓰기 모드)

f.write('Fine, thanks.\n') # 파일 포지션 마지막에 
f.write('And you?\n')

f.close()                   # 닫기
```

```python
# # 파일에서 모든 행의 문자열을 읽어들여 출력하기

f = open('hello.txt', encoding= "UTF-8")       # 열기(텍스트 + 읽기 모드)

while True:
    line = f.readline()
    if not line:            # 읽지 못했다(파일의 끝에 도달)
        break
    print(line, end='')

f.close()                   # 닫기
```

```python
# 파일에서 모든 행의 문자열을 읽어 들여 출력하기(read 메소드)

f = open('hello.txt', encoding="UTF-8")       # 열기(텍스트 + 읽기 모드)

lines = f.read()
print(lines, end='')

f.close()                   # 닫기
```

```python
# 파일에서 모든 행의 문자열을 읽어들여 출력하기(readlines 메소드)

f = open('hello.txt')       # 열기(텍스트 + 읽기 모드)

lines = f.readlines()
for line in lines:
    print(line, end='')

f.close()                   # 닫기
```

```python
# 파일에서 문자열을 읽어들여 출력하기(예외 처리)

try:
    f = open('hello.txt', 'r')
    try:
        for line in f: # 파일 읽기 기능 파일 자체가 iterator
            print(line, end='')
    except OSError:
        pass             # 읽기 실패 시 대처 방안
    finally:
        f.close()
except OSError:
    pass                 # 오픈 실패 시 대처 방안
```

```python
# 파일에서 문자열을 읽어들여 출력하기(with문)

with open('hello.txt', 'r') as f:
    for line in f:
        print(line, end='' )


# with 문은 종료 처리를 자동으로 수행
```
