```python
def operation_score(score):
    #implement your code
    sum_score = sum(score)
    avg_score = sum_score/len(score)

    return (sum_score, avg_score)

input_data_operation_score = [[32, 68, 72, 54, 92], [50, 5], [40,50,87,100]]
output_data_operation_score = [(318, 63.6), (55, 27.5), (277, 69.25)]

for input_values, output_value in zip(input_data_operation_score, output_data_operation_score):
    assert operation_score(input_values) == output_value
```

```python
# list comprehension

x = [n for n in range(1,8)]
print(x)

x = [n**2 for n in range(1,8)]
print(x)

x = [f'{n}{"+" if n % 2 == 1 else "-"}' for n in range(1, 8)]
print(x)

x = [n for n in range(1,8) if n % 2 == 0]
print(x)

x = [i * 10 + j for i in range(1,3) for j in range(1, 4)]
print(x)

table = [[i * 10 + j for i in range(1,3)] for j in range(1,4)]
print(table)

n=4
im = [[1 if i == j else 0 for i in range(n)] for j in range(n)]
print(im)
```
