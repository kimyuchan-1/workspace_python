```python
# 점수를 입력받아 총점과 평균 구하기 1

print('총점과 평균을 구해봅시다.')
number = int(input('학생 수：'))

score = [None] * number

for i in range(number):
    score[i] = int(input('{}번 학생의 점수：'.format(i + 1)))

total = 0
for i in range(number):
    total += score[i]

print('총점은 {}점입니다.'.format(total))
print('평균은 {}점입니다.'.format(total / number))
```

```python
# 점수를 입력받아 총점과 평균 구하기 2(오류)

print('총점과 평균을 구해봅시다.')
number = int(input('학생 수：'))

score = [None] * number

for i, point in enumerate(score):
    point = int(input('{}번 학생의 점수：'.format(i + 1)))
    score[i] = point

total = 0
for i in range(number):
    total += score[i]

print('총점은 {}점입니다.'.format(total))
print('평균은 {}점입니다.'.format(total / number))
```

```python
# 점수를 입력받아 총점과 평균 구하기 5(sum 함수로 총점 구하기)

print('총점과 평균을 구해봅시다.')
number = int(input('학생 수：'))

score = [None] * number

for i in range(number):
    score[i] = int(input('{}번 학생의 점수：'.format(i + 1)))

total = sum(score)

print('총점은 {}점입니다.'.format(total))
print('평균은 {}점입니다.'.format(total / number))
```

```python
# 학생 수 없이 점수만 입력받아 총점과 평균 구하기

print('총점과 평균을 구해봅시다.')
print('"End"를 입력하면 종료됩니다.')

number = 0
score = []                  # 빈 리스트

while True:
    s = input('{}번 학생의 점수：'.format(number + 1))
    if s == 'End':
        break
    score.append(int(s))    # 마지막에 추가
    number += 1

total = sum(score)

print('총점은 {}점입니다.'.format(total))
print('평균은 {}점입니다.'.format(total / number))


```

```python
# 학생 수 없이 점수만 입력받아 최고점과 최저점 구하기

print('최고점과 최저점을 구해봅시다.')
print('"End"를 입력하면 종료됩니다.')

number = 0
score = []                  # 빈 리스트

while True:
    s = input('{}번 학생의 점수：'.format(number + 1))
    if s == 'End':
        break
    score.append(int(s))    # 마지막에 추가
    number += 1

minimum = maximum = score[0]
for i in range(1, number):
    if score[i] < minimum: minimum = score[i]
    if score[i] > maximum: maximum = score[i]

print('최고점은 {}점입니다.'.format(maximum))
print('최저점은 {}점입니다.'.format(minimum))

```

```python
# 학생 수 없이 점수만 입력받아 최고점과 최저점 구하기 2(내장 함수 사용하기)

print('최고점과 최저점을 구해봅시다.')
print('"End"를 입력하면 종료됩니다.')

number = 0
score = []                  # 빈 리스트

while True:
    s = input('{}번 학생의 점수：'.format(number + 1))
    if s == 'End':
        break
    score.append(int(s))    # 마지막에 추가
    number += 1

minimum = min(score)
maximum = max(score)

print('최고점은 {}점입니다.'.format(maximum))
print('최저점은 {}점입니다.'.format(minimum))

```

```python
# 행렬의 합 구하기(행수/열수/값 입력받기)
import random

print('행렬의 합을 구해봅시다.')
height = int(input('행수:'))
width  = int(input('열수:'))

a = [None] * height
for i in range(height):
    a[i] = [None] * width
    for j in range(width):
        #a[i][j] = int(input('a[{}][{}] : '.format(i, j)))
        # 난수 생성
        a[i][j] = random.randint(1,10)


b = [None] * height
for i in range(height):
    b[i] = [None] * width
    for j in range(width):
        #b[i][j] = int(input('b[{}][{}] : '.format(i, j)))
         b[i][j] = random.randint(1,10)

c = [None] * height
for i in range(height):
    c[i] = [None] * width
    for j in range(width):
        c[i][j] = a[i][j] + b[i][j]

for i in range(height):
    for j in range(width):
        print('c[{}][{}] = {}'.format(i, j, c[i][j]))
```
