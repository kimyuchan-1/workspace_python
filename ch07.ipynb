```python
def operation_score(score):
    #implement your code
    sum_score = sum(score)
    avg_score = sum_score/len(score)

    return (sum_score, avg_score)

input_data_operation_score = [[32, 68, 72, 54, 92], [50, 5], [40,50,87,100]]
output_data_operation_score = [(318, 63.6), (55, 27.5), (277, 69.25)]

for input_values, output_value in zip(input_data_operation_score, output_data_operation_score):
    assert operation_score(input_values) == output_value
```
