```python
# 튜플 생성

tup = (1,2,3)
print(tup)

tup1 = tuple(range(5))
print(tup1)

tup2 = 1,2,3
print(tup2)

n = 1
t = 1,
print(type(n))
print(type(t))

tup3 = tuple([1,2,3])
print(tup3)

tup4 = divmod(13, 3)
print(tup4)
```

```python
tpl = (1,2,3)
print(id(tpl))
tpl += (4, 5)
print(tpl)
print(id(tpl))
```

```python
x = (1, 3, 2)
print(id(x))
x = tuple(sorted(x))
print(x)
print(id(x))
```

```python
tpl1 = (1,2,3,4,5)
tpl2 = tuple(tpl1)
print(id(tpl1))
print(id(tpl2))
```

```python
tuple4 = 1,2,3,4
a,b,c,d = tuple4
a,_,c,_ = tuple4

a,b,c,d
```
