```python
# 튜플 생성

tup = (1,2,3)
print(tup)

tup1 = tuple(range(5))
print(tup1)

tup2 = 1,2,3
print(tup2)

n = 1
t = 1,
print(type(n))
print(type(t))

tup3 = tuple([1,2,3])
print(tup3)

tup4 = divmod(13, 3)
print(tup4)
```

```python
tpl = (1,2,3)
print(id(tpl))
tpl += (4, 5)
print(tpl)
print(id(tpl))
```

```python
x = (1, 3, 2)
print(id(x))
x = tuple(sorted(x))
print(x)
print(id(x))
```

```python
tpl1 = (1,2,3,4,5)
tpl2 = tuple(tpl1)
print(id(tpl1))
print(id(tpl2))
```

```python
tuple4 = 1,2,3,4
a,b,c,d = tuple4
a,_,c,_ = tuple4

a,b,c,d
```

```python
a= 1
b=2
print(a,b)

a, b = b, a # tuple 개념이 있기 때문에 가능한 것
print(a,b)
```

```python
a, _, *c = tuple4
print(a,c)

a,b,c,*d,e = tuple4
print(a,b,c,d,e)
```

```python
tuple5 = (1,2,(3,4))
a,b,c = tuple5
print(a,b,c)
```

```python
import string

s = string.ascii_lowercase

for i, ch in enumerate(s):
    print(f's[{i}] = {ch}')
```

```python
slist = list(string.ascii_lowercase)

for idx, ch in enumerate(slist):
    print(f'slist[{idx}] = {ch}')
```

```python
p1 = [72, 56, 89]
p2 = [42, 85, 77]
name = ['김서준', '이서아', '최하준']
pt = tuple(zip(p1,p2,name))
print(pt)
print(pt[0])
print(pt[0][0]) # tuple의 요소를 선택할 때 인덱싱 사용
```
