```python
season = {
    '봄': 'spring',
    '여름': 'summer',
    '가을': 'autumn',
    '겨울': 'winter',
}
print(f'season["봄"]: {season['봄']}')
```

```python
# dictionary 생성
dict01= {'Korea' : 410, 'China': 156, "Japan": 392, "Taiwon" : 158}
print(dict01)
lst = [('Korea', 410), ('China', 156), ("Japan", 392), ("Taiwon", 158)]
dict02 = dict(lst)
print(dict02)
```

```python
dict1 = {'KOR':'Korea', 'CHN':'China','JPN':'Japan'}
dict2 = {'KOR':'Korea', 'JPN':'Japan','CHN':'China'}

print(dict1 == dict2)
# key-value mapping이 동일한 dictionary는 같은 구조로 판단

print(dict1['KOR']) # 인덱싱할 때 key값을 활용
```

```python
color = {'red':'빨강','green':'초록','blue':'파랑'}
print(color.get('red'))
# color['pink'] : 오류 발생
print(color.get('pink')) # 오류없이 default값인 None 출력
```

```python
# key값으로 value값 업데이트
dict1['KOR'] = 'Republic of Korea'
print(dict1)

# key값이 없으면 새로운 요소 추가
dict1['USA'] = 'United State of America'
print(dict1)
```

```python
# dictionary 출력(기존에 key값을 가진 경우) 및 추가(key값이 없는 경우)
v1 = color.setdefault('red','레드') # value값이 변경되진 않음
print(v1)
v2 = color.setdefault('yellow','노랑')
print(v2)
print(color)
```

```python
# update() : 딕셔너리 결합 + 기존 딕셔너리와 동일한 key값이 있는 경우 value값을 최신화 + 제자리 연산
cmy = {'cyan':'하늘','magenta':'자주','yellow':'황'}
color.update(cmy)
print(color)

# update()의 파라미터가 딕셔너리 형태가 아니더라도 수정 가능, key = 'value'
### 함수의 매개변수 전달 방식: 키워드 전달 방식
color.update(yellow = '노랑')
print(color)
```

```python
# dictionary 요소 제거
# pop() 매개변수로 key값을 받음
v1 = color.pop('yellow')
print(v1)
print(color)

# del 
del color['magenta']
print(color)
```

```python
# key in dictionary
if 'purple' not in color:
    color['purple'] = '보라'
print(color)

print(len(color))

b = color.copy()
print(b)
```

```python
# enumerate
for idx, key in enumerate(color):
    print(f'{idx}  {key} : {color[key]}')
```

```python
# keys(), values(), items() => dictionary view
print(f'key : {list(color.keys())}')
print(f'value : {list(color.values())}')
print(f'item : {list(color.items())}')
```

```python
txt = 'ABAXB'
count = {}
for ch in txt:
    if ch not in count:
        count[ch] = 1
    else:
        count[ch] += 1

print(count)
```

```python
# dictionary comprehension
count = {ch: txt.count(ch) for ch in txt}
print(count)
```

```python
s = ['봄','여름','가을','겨울']
season = {k:v for k,v in enumerate(s)}
print(season)
```
