```python
# set 생성
set01 = {1}
print(set01)
set02 = set([1,2,3,2])
print(set02)
```

```python
set3 = set([1,2,3,4])
print(id(set3))
set3.add(5)
# add(), remove(), pop(): 주소 변경 x / 다른 집합과의 연산: union(), intersection(), difference()를 통해서는 주소 변경
print(id(set3))
print(set3)
set3 = set3.union({4,5,6})
print(set3)
print(id(set3))
```

```python
## 1과 5가 같은 집합인지를 파악하는 방법
n = 0
m = 2
if n in set3 and m in set3:
    print("cycle")
```
