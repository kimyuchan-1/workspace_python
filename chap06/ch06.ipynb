```python
data = list(map(int, input().split()))
```

```python
data
```

```python
result = [i**2 for i in data]
result
```

```python
# 스페이스로 구분하여 한 줄에 출력
print(" ".join(map(str, result)))

# 콤마(,)로 구분하여 한 줄에 출력
print(",".join(map(str, result)))

# 엔터로 구분하여 여러 줄에 출력
print("\n".join(map(str, result)))
```
