## μ‹¤μµ λ¬Έμ : λ¶€λ™μ†μμ μ μ•μ „ν• λΉ„κµ ν•¨μ κµ¬ν„ν•κΈ°
**μ„¤λ…**: μ»΄ν“¨ν„°λ” μ‹­μ§„μ μ†μλ¥Ό μ΄μ§„μλ΅ λ³€ν™ν•λ” κ³Όμ •μ—μ„ λ―Έμ„Έν• μ¤μ°¨λ¥Ό κ°€μ§‘λ‹λ‹¤. μ΄λ΅ μΈν•΄ `0.1 + 0.2`λ” μ •ν™•ν `0.3`μ΄ μ•„λ‹λ―€λ΅ `==` μ—°μ‚°μλ΅ λΉ„κµν•λ©΄ `False`λ¥Ό λ°ν™ν•©λ‹λ‹¤. μ΄ λ¬Έμ λ¥Ό ν•΄κ²°ν•κΈ° μ„ν•΄, λ‘ μ«μμ μ°¨μ΄κ°€ μ•„μ£Ό μ‘μ€ κ°’(ν—μ© μ¤μ°¨, μ—΅μ‹¤λ΅ )λ³΄λ‹¤ μ‘μ€μ§€λ¥Ό ν™•μΈν•λ” λ°©λ²•μ΄ μ‚¬μ©λ©λ‹λ‹¤. λ‘ λ¶€λ™μ†μμ  μ«μλ¥Ό μ•μ „ν•κ² λΉ„κµν•λ” ν•¨μλ¥Ό μ§μ ‘ κµ¬ν„ν•΄ λ³΄μ„Έμ”.

$$ |a - b| < \epsilon $$

**μ”κµ¬μ‚¬ν•­**:
- `are_floats_equal`μ΄λΌλ” μ΄λ¦„μ ν•¨μλ¥Ό μ™„μ„±ν•μ—¬, λ‘ λ¶€λ™μ†μμ  `a`μ™€ `b`μ μ°¨μ΄(μ λ“κ°’)κ°€ μ£Όμ–΄μ§„ ν—μ© μ¤μ°¨ `epsilon`λ³΄λ‹¤ μ‘μ€μ§€ κ²€μ‚¬ν•μ„Έμ”.
- `a`μ— `0.1 + 0.2`λ¥Ό, `b`μ— `0.3`μ„ ν• λ‹Ήν•μ„Έμ”.
- μ™„μ„±λ ν•¨μλ¥Ό μ‚¬μ©ν•μ—¬ `a`μ™€ `b`κ°€ μ‹¤μ§μ μΌλ΅ κ°™μ€μ§€ ν™•μΈν•κ³  κ²°κ³Όλ¥Ό μ¶λ ¥ν•μ„Έμ”.



```python
def are_floats_equal(a: float, b: float, epsilon: float = 1e-9) -> bool:
    """
    λ‘ λ¶€λ™μ†μμ  μ«μκ°€ ν—μ© μ¤μ°¨(epsilon) λ‚΄μ—μ„ κ°™μ€μ§€ ν™•μΈν•©λ‹λ‹¤.

    Args:
        a (float): μ²« λ²μ§Έ λ¶€λ™μ†μμ  μ«μ.
        b (float): λ‘ λ²μ§Έ λ¶€λ™μ†μμ  μ«μ.
        epsilon (float): ν—μ©λλ” μµλ€ μ¤μ°¨. κΈ°λ³Έκ°’μ€ 1e-9μ…λ‹λ‹¤.

    Returns:
        bool: λ‘ μ«μκ°€ ν—μ© μ¤μ°¨ λ‚΄μ—μ„ κ°™μΌλ©΄ True, κ·Έλ ‡μ§€ μ•μΌλ©΄ False.
    """
    # μ΄ λ¶€λ¶„μ„ κµ¬ν„ν•μ„Έμ”.
    return abs(a - b) < epsilon
```

```python
# ν…μ¤νΈ μ½”λ“
num1 = 0.1 + 0.2
num2 = 0.3

# ν•¨μλ¥Ό νΈμ¶ν•μ—¬ κ²°κ³Όλ¥Ό ν™•μΈν•κ³  μ¶λ ¥ν•μ„Έμ”.
result = are_floats_equal(num1, num2)
print(f"0.1 + 0.2μ™€ 0.3μ€ κ°™μ€κ°€? {result}")
```

## μ‹¤μµ λ¬Έμ : λ¶€λ™μ†μμ  λ„μ  μ¤μ°¨ μ‹λ®¬λ μ΄μ…
**μ„¤λ…**: λ‹¨μΌ λ¶€λ™μ†μμ  μ—°μ‚°μ μ¤μ°¨λ” λ§¤μ° μ‘μ§€λ§, μ΄λ¬ν• μ‘μ€ μ¤μ°¨λ“¤μ΄ λ°λ³µλ¬Έ μ•μ—μ„ μμ², μλ§ λ² λ”ν•΄μ§€λ©΄ λ¬΄μ‹ν•  μ μ—†λ” **λ„μ  μ¤μ°¨(Accumulated Error)** λ΅ μ»¤μ§ μ μμµλ‹λ‹¤. μ΄λ² μ‹¤μµμ—μ„λ” `for` λ£¨ν”„λ¥Ό μ‚¬μ©ν•΄ μ‘μ€ μ†μ(`0.1`)λ¥Ό μ—¬λ¬ λ² λ”ν•λ” κ°„λ‹¨ν• μ‹λ®¬λ μ΄μ…μ„ ν†µν•΄, μ΄λ΅ μ μΈ κ²°κ³Όμ™€ μ‹¤μ  κ³„μ‚° κ²°κ³Ό μ‚¬μ΄μ— μ¤μ°¨κ°€ μ–΄λ–»κ² λ°μƒν•λ”μ§€ μ§μ ‘ ν™•μΈν•΄ λ³΄κ² μµλ‹λ‹¤. π’»

$$ \sum_{i=1}^{N} x \neq N \times x \quad (\text{in floating-point arithmetic}) $$

**μ”κµ¬μ‚¬ν•­**:
- `accumulate_error` ν•¨μ λ‚΄μ—μ„ `for` λ£¨ν”„λ¥Ό μ‚¬μ©ν•΄ `increment` κ°’μ„ `iterations` νμλ§νΌ `total`μ— λ”ν•λ” μ½”λ“λ¥Ό μ™„μ„±ν•μ„Έμ”.
- `0.1`μ„ `10000`λ² λ”ν• κ²°κ³Όλ¥Ό ν•¨μλ¥Ό νΈμ¶ν•μ—¬ μ–»μΌμ„Έμ”.
- μν•™μ μΌλ΅ μμƒλλ” κ²°κ³Ό(`0.1 * 10000`)μ™€ λ£¨ν”„λ¥Ό ν†µν•΄ μ–»μ€ μ‹¤μ  κ²°κ³Όλ¥Ό λΉ„κµν•κ³ , κ·Έ μ°¨μ΄(λ„μ  μ¤μ°¨)λ¥Ό μ¶λ ¥ν•μ—¬ ν™•μΈν•μ„Έμ”.


```python

def accumulate_error(increment: float, iterations: int) -> float:
    """
    μ£Όμ–΄μ§„ increment κ°’μ„ iterations νμλ§νΌ λ°λ³µν•΄μ„ λ”ν•μ—¬
    λ¶€λ™μ†μμ μ λ„μ  μ¤μ°¨λ¥Ό ν™•μΈν•©λ‹λ‹¤.

    Args:
        increment (float): κ° λ°λ³µλ§λ‹¤ λ”ν•  κ°’.
        iterations (int): λ°λ³µ νμ.

    Returns:
        float: λ„μ λ ν•©κ³„.
    """
    total = 0.0
    # for λ£¨ν”„λ¥Ό μ‚¬μ©ν•μ—¬ incrementλ¥Ό totalμ— iterations λ§νΌ λ”ν•μ„Έμ”.
    # μ΄ λ¶€λ¶„μ„ κµ¬ν„ν•μ„Έμ”.
    for i in range(iterations):
        total += increment

    return total

```

```python
# μ‹λ®¬λ μ΄μ… νλΌλ―Έν„°
inc = 0.1
iters = 10000

# ν•¨μ νΈμ¶ λ° κ²°κ³Ό κ³„μ‚°
actual_sum = accumulate_error(inc, iters)
expected_sum = inc * iters
error = actual_sum - expected_sum

print(f"λ£¨ν”„λ¥Ό μ΄μ©ν• κ³„μ‚° κ²°κ³Ό: {actual_sum}")
print(f"μν•™μ  μμƒ κ²°κ³Ό: {expected_sum}")
print(f"λ„μ  μ¤μ°¨: {error}")
```
